# Selezione parole per PARALLEL COORDINATES - plot!


```python
import os
import pandas as pd
import numpy as np
import json
```

```python
path = r'C:\Users\Attilio\Desktop\progettone\Disney_Animation_new'
```

```python
list_of_words = ['queen','papa','mama','dad','mr','miss','king', 'magic','ring','monster','forest','captain','kiss','car','land','silly','heart','sir','pop','evil','mad','dear','snow','battle','devil','scared','uncle','saving','brave','shoot','kill','trouble','dead','hero','dress','friend','school','god','weird','genius','space','crazy']
```

```python
len(list_of_words)
```

```python
list_of_max = []

for r, d, files in os.walk(path):
    for f in files:
        filename = r+'\\\\'+f
        anno = f.split('-')[0]
        with open(filename, 'r') as b:
            reader = b.readline()
            value = reader.split(',')[1]
            list_of_max.append(float(value))
MAX = max(list_of_max)                    
```

```python
MAX
```

```python
#le dieci parole piÃ¹ rappresentative di ogni epoca
list_of_dict = []

for r, d, files in os.walk(path):
    for f in files:
        filename = r+'\\\\'+f
        anno = f.split('-')[0]
        count = 0
        diz = {}
        diz1 = {}
        with open(filename, 'r') as b:
            diz['key'] = anno
            for x in b:
                if count >= 10:
                    break
                word_score = x.split(',')

                diz1[word_score[0]] = float(word_score[1])
                count += 1
                


            for k in diz1:
                diz1[k] = diz1[k]/MAX

            diz['value'] = diz1
            list_of_dict.append(diz)
```

```python
#evoluzione di tot parole nella disney
list_of_dict = []

for r, d, files in os.walk(path):
    for f in files:
        filename = r+'\\\\'+f
        anno = f.split('-')[0]

        diz = {}
        diz1 = {}
        with open(filename, 'r') as b:
            diz['key'] = anno
            for x in b:
                word_score = x.split(',')
                if word_score[0] in list_of_words:

                    diz1[word_score[0]] = float(word_score[1])


            for k in diz1:
                diz1[k] = diz1[k]/MAX

            diz['value'] = diz1
            list_of_dict.append(diz)
```

```python
list_of_filenames = []
for r, d, files in os.walk(path):
    for f in files:
        filename = r+'\\'+f
        list_of_filenames.append(filename)
```

```python
lista_di_dizionari = []
for word in list_of_words:
    dizionario = {}
    for filename in list_of_filenames:
        anno = filename.split('\\')[-1].split('-')[0]
        with open(filename, 'r') as b:
            for x in b:
                lista_parole = []
                word_score = x.split(',')
                lista_parole.append(word_score[0])
                if word == word_score[0]:
                    dizionario['word'] = word_score[0]
                    dizionario[anno] = round(float(word_score[1]),3)/MAX
                
    lista_di_dizionari.append(dizionario)
```

```python
lista_di_dizionari
```

```python
df_parallel = pd.DataFrame(lista_di_dizionari).fillna(0)
```

```python
df_parallel['word'].unique()
```

```python
#df_parallel.to_csv(r'C:\Users\Attilio\Desktop\progettone\df_parallel.csv', index=False)
```

```python
lista = []
count = 0
for x in df_parallel.values:
    dizionario = {
        'word_col': count/len(df_parallel.values),
        'word': x[0],
        '1930': x[1],
        '1940': x[2],
        '1950': x[3],
        '1960': x[4],
        '1970': x[5],
        '1980': x[6],
        '1990': x[7],
        '2000': x[8],
        '2010': x[9]
    }
    lista.append(dizionario)
    count += 1
```

```python
with open('data.json', 'w') as outfile:
    json.dump(lista, outfile)
```

```python
import plotly.express as px
fig = px.parallel_coordinates(df_parallel,
                              dimensions=['1930', '1940', '1950','1960','1970','1980','1990','2000','2010'],
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)
fig.show()
```
