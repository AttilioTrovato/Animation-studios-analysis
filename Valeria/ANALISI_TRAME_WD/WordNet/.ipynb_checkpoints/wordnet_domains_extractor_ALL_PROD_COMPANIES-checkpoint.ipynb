```python
import import_ipynb
import os
import pandas as pd
import gensim
from collections import Counter
from pprint import pprint

import nltk
from nltk.corpus import stopwords
en_stopwords = stopwords.words('english') + ['return']

import spacy
nlp = spacy.load('en')

from WNDomains import WordNetDomains 
wnd = WordNetDomains(wordnet_home='.')

import seaborn as sns; sns.set()
from matplotlib import pyplot as plt
```

```python
path= r'/Users/valeriaguttilla/Desktop/MASTER/PROGETTONE/ANALISI_TESTUALE/ANALISI TRAME/TRAME/ALL_COMPANIES/'


THRESHOLD = 0.4
dict_all_domains=dict()

for dirpath, dirnames, files in os.walk(path):    
    if len(files) < 1:
        continue
        
    house_domains = {}
    house_domain_counter = Counter()
    tot_house_movies = len(files)
    
    for file in files:
        if file[0] == '.':
            continue
        
        movie_domain_counter = Counter()

        with open(dirpath+ "/" + file, 'r', encoding='Latin-1') as f:
            text = f.read()
            doc = nlp(text)
            words = set([token.text for token in doc if (len(token.text) > 2 and token.pos_ in ['NOUN', 'PROPN'])])
            words = words.difference(en_stopwords) 

            movie=file.split('.')[0]
            house=dirpath.split('/')[-1]

            #print(movie)
            for word in words:
                domain = wnd.get_domains(word, 'n', first_sense_only=True)
                
                # since domain is a set()
                for tmp_d in domain:
                    if tmp_d != 'factotum':
                        #print(word, '->', tmp_d)
                        movie_domain_counter[tmp_d] += 1
                    
            for md in movie_domain_counter:
                if movie_domain_counter[md] > 1:
                    house_domain_counter[md] += 1
                
            #pprint(movie_domain_counter)
            
    #print(house, tot_house_movies)
    for dd in house_domain_counter:
        print(house, dd, house_domain_counter[dd])
        domain_movies_len = house_domain_counter[dd]
        
        if domain_movies_len >= (tot_house_movies * THRESHOLD):
            weight = domain_movies_len/(tot_house_movies*1.0)
            
            if dd in dict_all_domains:
                dict_all_domains[dd][house] = weight
            else:
                dict_all_domains[dd] = {house: weight}
                
```

```python
full_df = pd.DataFrame(columns=['Aardman', 'Blue', 'Disney', 'Dreamworks', 'Illumination', 'Paramount', 'Sony', 'StudioGhibli', 'Universal', 'Warner'])

for dom, houses in dict_all_domains.items():
    full_df.loc[dom] = [round(houses.get(str(house),0.0), 2) for house in ['AARDMAN', 'BLUE', 'DISNEY', 'DREAMWORKS', 'ILLUMINATION', 'PARAMOUNT', 'SONY', 'GHIBLI', 'UNIVERSAL', 'WARNER']]
    
full_df.sort_index(inplace=True)
```

```python
def to_flare_json(df, filename):
    """Convert dataframe into nested JSON as in flare files used for D3.js"""
    flare = dict()
    d = {"name":"flare", "children": []}
    
    
    for casa in df.index:
        for column in df.columns:
            parent=casa
            child = column
            child_size=df[child][parent]
            #print(parent, child, child_size)

        # Make a list of keys
            key_list = []
            for item in d['children']:
                key_list.append(item['name'])

            #if 'parent' is NOT a key in flare.JSON, append it
            if not parent in key_list:
                if child_size > 0.5:
                    d['children'].append({"name": parent, "children":[{"value": child_size, "name": child}]})
            # if parent IS a key in flare.json, add a new child to it
            else:
                if child_size > 0.5:
                    d['children'][key_list.index(parent)]['children'].append({"value": child_size, "name": child})
        flare = d
        # export the final result to a json file
    with open(filename +'.json', 'w') as outfile:
        json.dump(flare, outfile, indent=1)
    return ("Done")
```

```python
def create_entries(df):
    entries = []
    # Stopping case
    if df.shape[1] == 2:  # only 2 columns left
        for i in range(df.shape[0]):  # iterating on rows
            entries.append(
                {"Name": df.iloc[i, 0],
                 df.columns[-1]: df.iloc[i, 1]}
            )
    # Iterating case
    else:
        values = set(df.iloc[:, 0])  # Getting the set of unique values
        for v in values:
            entries.append(
                {"Name": v,
                 # reiterating the process but without the first column
                 # and only the rows with the current value
                 "Children": create_entries(
                     df.loc[df.iloc[:, 0] == v].iloc[:, 1:]
                 )}
            )
    return entries
```
