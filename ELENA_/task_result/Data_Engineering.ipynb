Normalization

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('text_analysis.csv')
```

```python
df.head()
```

```python
df.drop('Short_Phrases%', axis=1, inplace=True)
```

```python
df.head()
```

```python
scaler=MinMaxScaler()
```

```python
indexes=[c for c in df.columns if c not in ['Title', 'House', 'Category', 'Decade', 'Year']]
```

```python
df[indexes] = scaler.fit_transform(df[indexes])
```

```python
df.head()
```

```python
df.to_csv('normalized_texts.csv', index=False)
```
