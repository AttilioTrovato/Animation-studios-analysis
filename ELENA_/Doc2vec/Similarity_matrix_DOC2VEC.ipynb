```python
data=[]
with open(input_file, encoding='utf-8', newline='') as handle:
    reader = csv.reader(handle)
    for row in reader:
        data.append(TaggedDocument(re.split('[\W\d_]+',row[5].lower()),[row[0]], row[]))
```

```python
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
model= Doc2Vec(data, vector_size=150, window=10, min_count=5)
```

```python
movies=list(model.docvecs.doctags.keys())
```

```python
movies.remove('Title')
```

```python
movies[0]
```

```python
vocab={}
for movie in movies:
    if movie not in vocab:
        vocab[movie] = len(vocab)
```

```python
vocab
```

```python
reverse_vocab = {}
for movie, movieID in vocab.items():
    reverse_vocab[movieID] = movie
```

```python
reverse_vocab
```

```python
matrix=np.zeros(shape=(len(movies), len(vocab)))
```

```python
for index, key in enumerate(vocab):
    for index2, key2 in enumerate(vocab):
        matrix[index][index2] = model.docvecs.similarity(key, key2)
        
len(matrix)
```

```python
matrix
```

```python
nodes=set()
links=[]
```

```python
for x in range(len(matrix)):
    for y in range(x+1, len(matrix)):
        triplet=(reverse_vocab[x],reverse_vocab[y], matrix[x][y])
        links.append(triplet)
        nodes |= set([reverse_vocab[x], reverse_vocab[y]])
                   
print('Nodes', nodes)
print('Links', links)
```

```python
graph = dict()
graph["nodes"] = [{"id": t} for t in nodes]
graph["links"] = [{"source":t[0], "target":t[1], "value":float(t[2])} for t in links]
with open('data_full.json', 'w') as outfile:
    json.dump(graph, outfile)
```
